#### import packages

```python
# import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# the first distribution
```

#### the first distribution - similar means, highly correlated

```python
# set seed
np.random.seed(123)

# parameters
dist1_mean = 50
dist1_std_dev = 10
dist1_corr = 0.8 # can change, but we will stick with 50, 10, corr = 0.8
dist1_cov_matrix = [[dist1_std_dev**2, dist1_corr * dist1_std_dev**2],
              [dist1_corr * dist1_std_dev**2, dist1_std_dev**2]]

# generate the data
data_one = np.random.multivariate_normal([dist1_mean, dist1_mean], dist1_cov_matrix, size=1000)
var_x_dist1 = data_one[:, 0] # create the figure
var_y_dist1 = data_one[:, 1]
plt.figure(figsize=(8, 6))
plt.scatter(var_x_dist1, var_y_dist1, alpha=0.5)
plt.title('Similar Means, Highly Correlated')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

#### the second distribution - similar means, uncorrelated

```python
# the second distribution
np.random.seed(123)
dist2_mean = 50  # need to tweak the parameters if you tweak above (or below)
dist2_std_dev = 10
var_x_dist2 = np.random.normal(dist2_mean, dist2_std_dev, size=1000)
var_y_dist2 = np.random.normal(dist2_mean, dist2_std_dev, size=1000)
plt.figure(figsize=(8, 6))
plt.scatter(var_x_dist2, var_y_dist2, alpha=0.5)
plt.title('Similar Means, Uncorrelated')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

#### the third distribution - different means, highly correlated

```python
# the third distribution
np.random.seed(123)
dist3_mean1 = 40
dist3_mean2 = 60
dist3_std_dev = 10
dist3_corr = 0.8 # basically same parameters, slightly altered
cov_matrix = [[dist3_std_dev**2, dist3_corr * dist3_std_dev**2],
              [dist3_corr * dist3_std_dev**2, dist3_std_dev**2]]
data_three = np.random.multivariate_normal([dist3_mean1, dist3_mean2], cov_matrix, size=1000) # the distribution

# separate the data
var_x = data_three[:, 0]
var_y = data_three[:, 1]
plt.figure(figsize=(8, 6))
plt.scatter(var_x, var_y, alpha=0.5)
plt.title('Different Means, Highly Correlated')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

#### the final distribution - different means, uncorrelated

```python
# the final distribution
np.random.seed(123)
dist4_mean1 = 40
dist4_mean2 = 60
std_dev = 10

# drawing from two different distributions
var_x = np.random.normal(dist4_mean1, std_dev, size=1000)
var_y = np.random.normal(dist4_mean2, std_dev, size=1000)

# the chart
plt.figure(figsize=(8, 6))
plt.scatter(var_x, var_y, alpha=0.5)
plt.title('Different Means, Uncorrelated')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()

```

#### graphing everything together

```python
# combine all four of these charts into one chart
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# each chart should be on a separate subplot
axes[0, 0].scatter(data_one[:, 0], data_one[:, 1], alpha=0.5)
axes[0, 0].set_title('Similar Means, Highly Correlated')
axes[0, 0].set_xlabel('Variable X')
axes[0, 0].set_ylabel('Variable Y')
axes[0, 0].grid(True)

axes[0, 1].scatter(var_x_dist2, var_y_dist2, alpha=0.5)
axes[0, 1].set_title('Similar Means, Uncorrelated')
axes[0, 1].set_xlabel('Variable X')
axes[0, 1].set_ylabel('Variable Y')
axes[0, 1].grid(True)

axes[1, 0].scatter(data_three[:, 0], data_three[:, 1], alpha=0.5)
axes[1, 0].set_title('Different Means, Highly Correlated')
axes[1, 0].set_xlabel('Variable X')
axes[1, 0].set_ylabel('Variable Y')
axes[1, 0].grid(True)

axes[1, 1].scatter(var_x, var_y, alpha=0.5)
axes[1, 1].set_title('Different Means, Uncorrelated')
axes[1, 1].set_xlabel('Variable X')
axes[1, 1].set_ylabel('Variable Y')
axes[1, 1].grid(True)

# spacing and display
plt.tight_layout()
plt.show()
```
